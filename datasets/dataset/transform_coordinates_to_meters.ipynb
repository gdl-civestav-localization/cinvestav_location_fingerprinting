```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
%load_ext autoreload
# %autoreload
from geopy.distance import vincenty
```

```python
df = pd.read_csv(
    'cinvestav_testbed.csv', 
    sep=',',
    engine='c'
)

df = df.drop('Unnamed: 0', 1)

df.head(5)
```

```python
x_0 = df['result_x'].min() # .median() # .mean()
y_0 = df['result_y'].min() # .median() # .mean()
print x_0, y_0

def distance_x(x):
    p1 = (x_0, y_0)
    p2 = (x, y_0)
    return vincenty(p1, p2).meters

def distance_y(y):
    p1 = (x_0, y_0)
    p2 = (x_0, y)
    return vincenty(p1, p2).meters


df['result_x'] = df.apply(
    lambda x: distance_x(
        x['result_x']
    ),
    axis=1
)

df['result_y'] = df.apply(
    lambda x: distance_y(
        x['result_y']
    ),
    axis=1
)

df#.to_csv('d.csv')
```
